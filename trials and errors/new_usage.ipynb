```python
#str[tracks['track', 'genre_top']].map(lambda genres: 1235 in genres)
#print(tracks['track', 'genre_top'].shape)
tracks['track', 'genre_top'] = tracks['track', 'genre_top'].replace(np.nan, 'NaN', regex=True)
tracks['track', 'genre_top'].map(lambda x: type(x)).head()
#tracks['track', 'genre_top']
#genres['title']
```

```python
small = tracks['set', 'subset'] <= 'small'
#print(tracks['track', 'genre_top'].shape)
#print(genres)

# return true/false
def select(genre):
    return tracks['track', 'genre_top'].map(lambda genres: genre in genres)


# map(lambda genres: genre in genres) return true or false on a vector; tracks['track', 'genre_top'] is a vector
# genre in tracks corresponding to features.
genre1 = select('Instrumental')  # Instrumental. 1235
genre2 = select('Hip-Hop')  # Hip-Hop. 21

# subset: small; genre: Instrumental or Hip-Hop, and mfcc.
X = features.loc[small & (genre1 | genre2), 'mfcc']
# Fit the model with X and apply the dimensionality reduction on X.
X = skl.decomposition.PCA(n_components=3).fit_transform(X)

# Encode labels with value between 0 and n_classes-1.
y = tracks.loc[small & (genre1 | genre2), ('track', 'genre_top')]
y = skl.preprocessing.LabelEncoder().fit_transform(y)

plt.scatter(X[:,0], X[:,1], X[:,2], c=y, cmap='RdBu', alpha=0.5)
plt.legend(['PCA1', 'PCA2'], ncol = 2, loc= 'lower right')
X.shape, y.shape
plt.savefig('PCA.pdf')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs=X[:,0], ys = X[:,1], zs=X[:,2], zdir='z', c=y,  cmap='winter')
ax.set_xlim(-300, 550)
ax.set_ylim(-600, 250)
ax.set_zlim(-100, 170)
#plt.show()
plt.savefig('PCA.jpg')
```

## 5 Genre classification

### 5.1 From features

```python
#small
small = tracks['set', 'subset'] <= 'small'

train = tracks['set', 'split'] == 'training'
val = tracks['set', 'split'] == 'validation'
test = tracks['set', 'split'] == 'test'

y_train = tracks.loc[small & train, ('track', 'genre_top')]
y_test = tracks.loc[small & test, ('track', 'genre_top')]
enc = skl.preprocessing.LabelEncoder()
y_train = enc.fit_transform(y_train)
y_test = enc.transform(y_test)

X_train = features.loc[small & train, 'mfcc']
X_test = features.loc[small & test, 'mfcc']

print('{} training examples, {} testing examples'.format(y_train.size, y_test.size))
print('{} features, {} classes'.format(X_train.shape[1], np.unique(y_train).size))

# Accuracy: 46.93%
```

```python
# medium
medium = tracks['set', 'subset'] <= 'medium'

train = tracks['set', 'split'] == 'training'
val = tracks['set', 'split'] == 'validation'
test = tracks['set', 'split'] == 'test'

y_train = tracks.loc[medium & train, ('track', 'genre_top')]
y_test = tracks.loc[medium & test, ('track', 'genre_top')]
enc = skl.preprocessing.LabelEncoder()
y_train = enc.fit_transform(y_train)
y_test = enc.transform(y_test)

X_train = features.loc[medium & train, 'mfcc']
X_test = features.loc[medium & test, 'mfcc']

print('{} training examples, {} testing examples'.format(y_train.size, y_test.size))
print('{} features, {} classes'.format(X_train.shape[1], np.unique(y_train).size))

# Accuracy: 60.89%
```

```python
# large
large = tracks['set', 'subset'] <= 'large' # level[0]: 'set', level[1]: 'subset';

train = tracks['set', 'split'] == 'training'
val = tracks['set', 'split'] == 'validation'
test = tracks['set', 'split'] == 'test'

y_train = tracks.loc[large & train, ('track', 'genre_top')]
y_test = tracks.loc[large & test, ('track', 'genre_top')]
enc = skl.preprocessing.LabelEncoder()
y_train = enc.fit_transform(y_train)
y_test = enc.transform(y_test)

X_train = features.loc[large & train, 'mfcc']
X_test = features.loc[large & test, 'mfcc']

print('{} training examples, {} testing examples'.format(y_train.size, y_test.size))
print('{} features, {} classes'.format(X_train.shape[1], np.unique(y_train).size))

# Accuracy: 62.23%
```

```python
# Be sure training samples are shuffled.
X_train, y_train = skl.utils.shuffle(X_train, y_train, random_state=42)

# Standardize features by removing the mean and scaling to unit variance.
scaler = skl.preprocessing.StandardScaler(copy=False)
scaler.fit_transform(X_train)
scaler.transform(X_test)

# Support vector classification.
clf = skl.svm.SVC()
clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
print('Accuracy: {:.2%}'.format(score))
```

### PCA-SVM

* with medium subset

```python
medium = tracks['set', 'subset'] <= 'medium'

train = tracks['set', 'split'] == 'training'
val = tracks['set', 'split'] == 'validation'
test = tracks['set', 'split'] == 'test'

y_train = tracks.loc[medium & train, ('track', 'genre_top')]
y_train = skl.preprocessing.LabelEncoder().fit_transform(y_train)

y_test = tracks.loc[medium & test, ('track', 'genre_top')]
y_test = skl.preprocessing.LabelEncoder().fit_transform(y_test)

X_train = features.loc[medium & train,:]
X_train = skl.decomposition.PCA(n_components=2).fit_transform(X_train)
X_test = features.loc[medium & test,:]
X_test = skl.decomposition.PCA(n_components=2).fit_transform(X_test)

print('{} training examples, {} testing examples'.format(y_train.size, y_test.size))
print('{} features, {} classes'.format(X_train.shape[1], np.unique(y_train).size))

# Accuracy: 60.89%
```

```python
# Be sure training samples are shuffled.
X_train, y_train = skl.utils.shuffle(X_train, y_train, random_state=42)

# Standardize features by removing the mean and scaling to unit variance.
scaler = skl.preprocessing.StandardScaler(copy=False)
scaler.fit_transform(X_train)
scaler.transform(X_test)

# Support vector classification.
clf = skl.svm.SVC(kernel="linear")
clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
score_train = clf.score(X_train, y_train)
print('Test Accuracy: {:.2%}'.format(score))
print('Train Accuracy: {:.2%}'.format(score_train))
```
